```python
import pandas as pd
```

```python
df= pd.read_excel("C:\\Users\\saith\\OneDrive\\Desktop\\Book1.xlsx" )
df
```

```python

# Group by year and calculate the average score
average_score_yearwise = df.groupby('year')['marks_scored'].mean()

print(average_score_yearwise)
```

```python

# Group by semester and calculate the average score
average_score_semsterwise = df.groupby('semster')['marks_scored'].mean()

print(average_score_semsterwise)
```

```python

# Group by subdivision and calculate the average score
average_score_subdivisionwise = df.groupby('subdivision')['marks_scored'].mean()

print(average_score_subdivisionwise)
```

```python


# Define a function to categorize scores
def categorize_marks(marks):
    if marks >= 90:
        return "Excellent"
    elif marks >= 80:
        return "Very Good"
    elif marks >= 70:
        return "Good"
    else:
        return "Average"

# Add a new column 'Category' using the apply function
df['Category'] = df['marks_scored'].apply(categorize_marks)

print(df)
```

```python

# Group by semester and calculate the statistics
semester_wise_score = df.groupby('semster')['marks_scored'].agg(['mean', 'median', 'std'])

print(semester_wise_score)
```

```python
import matplotlib.pyplot as plt


```

```python
# Group by semester and calculate the average marks
semester_average = df.groupby('semster')['marks_scored'].mean()

# Create the line plot
plt.plot(semester_average.index, semester_average.values, marker='o')

# Set the labels and title
plt.xlabel('Semester')
plt.ylabel('Average Marks')
plt.title('Semester-wise Average Marks')

# Display the plot
plt.show()
```

```python


# Group by semester and calculate the average marks
semester_average = df.groupby('semster')['marks_scored'].mean()

# Create the bar chart
plt.bar(semester_average.index, semester_average.values)

# Set the labels and title
plt.xlabel('Semester')
plt.ylabel('Average Marks')
plt.title('Semester-wise Average Marks')
plt.ylim(73.5, 76)

# Display the plot
plt.show()
```

```python


# Create a histogram
plt.figure(figsize=(14, 6))  # Set the size of the figure
plt.hist(df['marks_scored'], bins=15, color='b' , alpha=0.7)  # Create the histogram
plt.xlabel('Marks')
plt.ylabel('Frequency')
plt.title('Distribution of Marks')

# Display the histogram
plt.show()

# Create a box plot
plt.figure(figsize=(6, 6))  # Set the size of the figure
plt.boxplot(df['marks_scored'], notch=False, showmeans=True, meanline = True, widths = 0.3 )  # Create the box plot
plt.xlabel('Marks')
plt.title('Box Plot of Marks')

# Display the box plot
plt.show()
```
